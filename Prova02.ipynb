```python
def selection_sort(arr):
    for i in range(len(arr)-1):
        min_idx = i + 1
        for j in range(i, len(arr)):
            if arr[j] < arr[i]:
                min_idx = j
        arr[min_idx] = arr[i]
        arr[i] = arr[min_idx]
    return arr
```

```python
# teste 1

lista = [29,10,14,37,13,5,11]

print (selection_sort(lista))
```

No primeiro teste, o algoritmo não funcionou. Para resolver vou analisar cada linha:


1ª linha: percorre a lista desde o primeiro elemento até o penultimo elemento -> está correto

2º linha: define como o menor da lista como o segundo elemento (indice 1) -> está errado, para comparar é preciso que o primeiro elemento da lista seja definido com min

3ª linha: faz o segundo loop para verificar se a variavél j é menor que o loop  i, porém está incorreta: j precisa ser i+ 1

4ª linha compara se j é menor que i, dessa forma está errada, pois compara sempre com um unico nº sem atualizar o correto é arr[min_idx]

5ª e 6ª da forma que está, etsá errada, pois a informação pode se perder. O correto é ter uma variável para guardar a informação antes de fazer a troca


```python
def selection_sort2(arr):
    for i in range(len(arr) - 1):
        min_idx = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]            
    return arr

```

```python
#lista aleatória
import random

lista2 = [random.randint(1, 10) for _ in range(7)]

print (lista2)

print(selection_sort2(lista2))
```
